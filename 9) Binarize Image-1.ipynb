```python
import cv2
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (25,10)
```

```python
img_name = 'diagram1.jpg'
img = cv2.imread('Data/'+ img_name, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='Greys_r')
plt.show()
```

```python
th1 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
```

```python
plt.imshow(th1, cmap='Greys_r')
plt.show()

plt.imshow(th2, cmap='Greys_r')
plt.show()
```
